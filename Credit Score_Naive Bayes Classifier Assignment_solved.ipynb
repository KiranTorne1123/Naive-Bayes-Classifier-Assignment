```python
# Segregating the Dataset into Input(x) and Output(y)

x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
x.shape, y.shape
```

```python
# Splitting the Dataset into Training and Testing Data
```

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('Importing is done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=33)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Loading the Gaussian Naive Bayes classifier model
```

```python
# We import additional libraries

from sklearn.naive_bayes import GaussianNB
print('Importing is done')
```

```python
gnb = GaussianNB()
gnb
```

```python
# We fit the training data in our model

gnb.fit(x_train, y_train)
gnb
```

```python
# Now testing new Model prediction by providing Test Data set

y_pred = gnb.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Calculating the Accuracy of the Trained Models
```

```python
# We import additional libraries

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print('Importing is done')
```

```python
# Now we check for accuracy of test data vs training data

accuracy = accuracy_score(y_test, y_pred)
print('Model accuracy : {0:0.3f}'. format (accuracy))
np.round(accuracy,2)*100
```

```python
# Generate a detailed classification report

report = classification_report(y_test, y_pred)
print(report)
```

```python
# We can see that our model executes with 100 % accuracy for this dataset.
```

```python
# Generate confusion matrix 

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# If we keep split as 70/30
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=33)
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
np.round(accuracy,2)*100
```

```python
# The results are still the same even with 70/30 split.
```
