```python
# 1. Line Chart for Temperature Over Time

# Convert date_custom to datetime format
df['date_custom'] = pd.to_datetime(df['date_custom'], format='%Y%m%d')

plt.figure(figsize=(12, 6))
plt.plot(df['date_custom'], df['temperature'], marker='o', linestyle='-')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# The above chart shows temperature variations taking place on each day during the mentioned tenure of readinds in dataset.
# Min = 14 & max = 24.
```

```python
# 2. Bar Chart for Rainfall Over Time

plt.figure(figsize=(12, 6))
plt.bar(df['date_custom'].dt.strftime('%Y-%m-%d'), df['rainfall'])
plt.title('Rainfall Over Time')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# The above chart displays the rainfall measured for each day over course of time of readings.
# Min recorded is 2 mm on 12th Jan 2022
# max recorded is 22 mm on 30th Jan 2022
```

```python
# 3. Scatter Plot for Temperature vs. Humidity

plt.figure(figsize=(10, 6))
plt.scatter(df['temperature'], df['humidity'], c='blue', marker='o')
plt.title('Temperature vs. Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# The above chart displays an inverse relation between humidity and temperature.
# Humidity decreases as the temperature increases.
# Humidity : under 30% is too dry, and over 50% is too high
```

```python
# 4. Histogram for Wind Speed

plt.figure(figsize=(10, 6))
plt.hist(df['wind_speed'], bins=10, color='green', edgecolor='black')
plt.title('Distribution of Wind Speed')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# The above chart shows recorded wind speeds 
# wind speed of 6 to 8 m/s is more often seen during the period of taking the records each day for the entire tenure.
```

```python
# 5. Box Plot for Multiple Variables

plt.figure(figsize=(12, 6))
df[['rainfall', 'temperature', 'humidity', 'wind_speed']].boxplot()
plt.title('Box Plot of Weather Variables')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# We can see that the variations in values are Humidity > rainfall > wind speed > temperature.
```

```python
# 6. Pie Chart for Weather Conditions

plt.figure(figsize=(8, 8))
weather_counts = df['weather_condition'].value_counts()
plt.pie(weather_counts, labels=['Rainy', 'Sunny'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Weather Conditions')
plt.show()
```

```python
# From all the readings in the dataset, it can be seen that the weather conditions was Rainy > Sunny. (79.2 % : 20.8 %)
```

```python
# 7. Correlation matrix
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# We can confirm that there is co-relation between temperature, humidity, wind speed VS weather condition.
```

```python
# Pairplot to visualize relationships of all features.

sns.pairplot(df, hue='weather_condition', markers=["o", "s"])
plt.suptitle('Pairplot of Features Colored by weather_condition', y=1.02)
plt.show()
```

```python
# Segregating the Dataset into Input(x) and Output(y)

x=df.iloc[:,:-2].values
x
```

```python
y=df['weather_condition'].values
y
```

```python
x.shape, y.shape
```

```python
# Splitting the Dataset into Training and Testing Data
```

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('Importing is done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=33)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Loading the Gaussian Naive Bayes classifier model
```

```python
# We import additional libraries

from sklearn.naive_bayes import GaussianNB
print('Importing is done')
```

```python
gnb = GaussianNB()
gnb
```

```python
# We fit the training data in our model

gnb.fit(x_train, y_train)
gnb
```

```python
# Now testing new Model prediction by providing Test Data set

y_pred = gnb.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Calculating the Accuracy of the Trained Models
```

```python
# We import additional libraries

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print('Importing is done')
```

```python
# Now we check for accuracy of test data vs training data

accuracy = accuracy_score(y_test, y_pred)
print('Model accuracy : {0:0.3f}'. format (accuracy))
np.round(accuracy,2)*100
```

```python
# Generate a detailed classification report

report = classification_report(y_test, y_pred)
print(report)
```

```python
# We can see that our model executes with 100 % accuracy for this dataset.
```

```python
# Generate confusion matrix 

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# If we keep split as 70/30
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=33)
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
np.round(accuracy,2)*100
```

```python
# The results are still the same even with 70/30 split.
```
