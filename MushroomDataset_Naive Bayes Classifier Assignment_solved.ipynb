```python
# Bar chart of all features versus class (target column)

# Convert 'class' to string for better readability in the plot
df['class'] = df['class'].astype(str)

# List of features to plot
features = ['cap-diameter', 'cap-shape', 'gill-attachment', 'gill-color', 'stem-height', 'stem-width', 'stem-color', 'season']

# Plotting bar plots for each feature versus class
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, hue='class', data=df, palette='viridis')
    plt.title(f'Bar Plot of {feature.replace("-", " ").capitalize()} vs Class')
    plt.xlabel(feature.replace("-", " ").capitalize())
    plt.ylabel('Count')
    plt.legend(title='Class')
    plt.show()
```

```python
# The above chart helps us understand the quality of mushroom produce falling under class 1 or 0, depending upon their characeristics.
```

```python
# Pie chart for total contribution of each class in the dataset.


def plot_pie_chart(feature, df):
    data = df[feature].value_counts()
    plt.figure(figsize=(8, 8))
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(data)))
    plt.title(f'Pie Chart of {feature.replace("-", " ").capitalize()}')
    plt.show()

plot_pie_chart('class', df)
```

```python
# From the above we can confirm that 54.9 % of the produce falls in class 1 and 45.1% falls in class 0.
```

```python
# Histogram of all features versus class.

# Convert 'class' to string for better readability in the plot
df['class'] = df['class'].astype(str)

# List of numerical features to plot
numerical_features = ['cap-diameter', 'cap-shape', 'gill-attachment', 'gill-color', 'stem-height', 'stem-width', 'stem-color', 'season']

# Plotting histograms for each numerical feature versus class
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=feature, hue='class', multiple="stack", palette='viridis', kde=True)
    plt.title(f'Histogram of {feature.replace("-", " ").capitalize()} vs Class')
    plt.xlabel(feature.replace("-", " ").capitalize())
    plt.ylabel('Count')
    plt.legend(title='Class')
    plt.show()
```

```python
# Segregating the Dataset into Input(x) and Output(y)

x=df.drop('class', axis=1).values
x
```

```python
y=df['class'].values
y
```

```python
# Scaling the features of the dataset
```

```python
# We import additional libraries

from sklearn.preprocessing import StandardScaler
print('Importing is done')
```

```python
scaler = StandardScaler()
scaler
```

```python
# Splitting the Dataset into Training and Testing Data
```

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('Importing is done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=33)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Loading the Gaussian Naive Bayes classifier model
```

```python
# We import additional libraries

from sklearn.naive_bayes import GaussianNB
print('Importing is done')
```

```python
gnb = GaussianNB()
gnb
```

```python
# We fit the training data in our model

gnb.fit(x_train, y_train)
gnb
```

```python
# Now testing new Model prediction by providing Test Data set

y_pred = gnb.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Calculating the Accuracy of the Trained Models
```

```python
# We import additional libraries

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print('Importing is done')
```

```python
# Now we check for accuracy of test data vs training data

accuracy = accuracy_score(y_test, y_pred)
print('Model accuracy : {0:0.3f}'. format (accuracy))
np.round(accuracy,2)*100
```

```python
# Generate a detailed classification report

report = classification_report(y_test, y_pred)
print(report)
```

```python
# We can see that our model executes with a 65% accuracy for this dataset.
```

```python
# Generate confusion matrix 

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# If we keep split as 70/30
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=33)
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
np.round(accuracy,2)*100
```
